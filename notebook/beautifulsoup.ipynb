```python
from bs4 import BeautifulSoup
from urllib import request
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:23.0) Gecko/20100101 Firefox/23.0'}  
req = request.Request(url="https://artist.artron.net/class-0-0-0.html", headers=headers)  
html = request.urlopen(req).read().decode('utf-8')
html
```

```python
soup = BeautifulSoup(html, features='lxml')
```

```python
soup.title
```

```python
artlist = soup.find_all('div',{"class":"shopList"})
artlist
```

```python
len(artlist)
```

```python
artlist_h3_list = []
for item in artlist:
    h3_str = item.find_all('a',{"class":"fl"})
    for h3 in h3_str:
        print(h3.get_text().strip())
        print(h3['href'])
```

```python
import re
```

```python
course_links = soup.find_all('a', {'href': re.compile('https://(.+?).artron.net'),'target':'_blank'})
for link in course_links:
    print(link['href'])
```
