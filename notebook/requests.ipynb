### requests: an alternative to urllib

```python
import requests
import webbrowser
```

```python
param = {"wd": "猎户座"}
r = requests.get('http://www.baidu.com/s', params=param)
print(r.url)
webbrowser.open(r.url) #打开浏览器访问r.url
```

```python
data = {'wb': '猎户座'}
headers = {'content-type': 'charset=utf-8'}
r = requests.post('http://www.baidu.com/s', data=data,headers=headers)
r.encoding='utf-8' 
print(r.text)
```

```python
IMAGE_URL = "https://scikit-learn.org/stable/_images/sphx_glr_plot_cluster_comparison_001.png"
```

```python
import requests
r = requests.get(IMAGE_URL)
with open('./img/image2.png', 'wb') as f:
    f.write(r.content)                      # whole document
```

## download chunck by chunck
Set stream = True in get() function. This is more efficient.

```python
r = requests.get(IMAGE_URL, stream=True)    # stream loading

with open('./img/image3.png', 'wb') as f:
    for chunk in r.iter_content(chunk_size=32):
        f.write(chunk)
```
